```python
tellar_file_name = './positions.txt'
bh_file_name = './positions_black_hole.txt'
generated_particles = './num_particles_generated.txt'

def update(i, x1,y1,z1,x2,y2,z2,x3,y3,z3):
    '''
    Parameters
    ---------------
    xi : ndarray 
        x-coordinates for each timestep
    yi : scalar
        y-coordinates for each timestep
    zi : scalar
        z-coordinates for each timestep

    
    Returns
    ---------------
    Plot containing all the coordinates for the the corrsponging timestep, to reproduce the 
    animation.
    '''
    sct1.set_data(x1[i], y1[i])
    sct1.set_3d_properties(z1[i])
    sct2.set_data(x2[i], y2[i])
    sct2.set_3d_properties(z2[i])
    sct3.set_data(x3[i], y3[i])
    sct3.set_3d_properties(z3[i])
    
    
    
    
with open(generated_particles, 'r') as f:
    data = f.readlines() 
for raw_line in data:
    split_line = raw_line.strip().split(',')
    n = [int(x) for x in split_line]
    
    
pos_gif = read_txt(stellar_file_name)
pos_gif_cntr = read_txt(bh_file_name)




x_1_gif = [[pos_gif[j][i][0] for i in range(0,n[0])] for j in range(len(pos_gif))]
y_1_gif = [[pos_gif[j][i][1] for i in range(0,n[0])] for j in range(len(pos_gif))]
z_1_gif = [[pos_gif[j][i][2] for i in range(0,n[0])] for j in range(len(pos_gif))]

x_2_gif = [[pos_gif[j][i][0] for i in range(n[0],len(pos_gif[0]))] for j in range(len(pos_gif))]
y_2_gif = [[pos_gif[j][i][1] for i in range(n[0],len(pos_gif[0]))] for j in range(len(pos_gif))]
z_2_gif = [[pos_gif[j][i][2] for i in range(n[0],len(pos_gif[0]))] for j in range(len(pos_gif))]

x_cntr_gif = [[pos_gif_cntr[j][i][0] for i in range(0,2)] for j in range(len(pos_gif_cntr))]
y_cntr_gif = [[pos_gif_cntr[j][i][1] for i in range(0,2)] for j in range(len(pos_gif_cntr))]
z_cntr_gif = [[pos_gif_cntr[j][i][2] for i in range(0,2)] for j in range(len(pos_gif_cntr))] 


fig = plt.figure(figsize=(7,7), constrained_layout=True)

ax = fig.add_subplot(111,projection='3d')


sct1, = ax.plot([], [], [], "o", markersize=1,color = 'blue')
sct2, = ax.plot([], [], [], "o", markersize=1,color = 'orange')
sct3, = ax.plot([], [], [], "o", markersize=2,color = 'black')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

ax.set_xlim([-500,500])
ax.set_ylim([-500,500])
ax.set_zlim([-800,200])

ani = animation.FuncAnimation(fig, update, \
fargs=(x_1_gif,y_1_gif,z_1_gif,x_2_gif,y_2_gif,z_2_gif,x_cntr_gif,y_cntr_gif,z_cntr_gif), \
                              interval=100)

plt.show()
```
